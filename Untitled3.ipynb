**Title of project:- Bank Customer Churn Model**

**Objective:-
The goal of this project is to predict whether a bank customer will churn (i.e., leave the bank). By identifying customers at risk of churning, the bank can take proactive measures to retain these customers.**

Data Source:-
YBI foundation

Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

Import Data

```python
data=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')
```

Describe Data

```python
data.head()
```

Data Visualization

```python
sns.countplot(x='Churn', data=data)  # Distribution of target variable

```

```python
sns.pairplot(data)
```

Data Preprocessing


```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
scaler = StandardScaler()
```

```python
scaled_features = scaler.fit_transform(data.drop('Churn', axis=1))
```

```python
scaled_data = pd.DataFrame(scaled_features, columns=data.columns[:-1])
```

Define Target Variable (y) and Feature Variables (X)

```python
X = scaled_data
y = data['Churn']
```

Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Modeling

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)

```

```python

model.fit(X_train, y_train)
```

Model Evaluation

```python
y_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
